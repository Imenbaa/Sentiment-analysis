```python
import csv
import pandas as pd
from youtube_transcript_api import YouTubeTranscriptApi
import os
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import words
```

```python
#L'OREAL INFALLIBLE PRO MATTE LIQUID LIPSTICK LES CHOCOLAT
lipstick=["Te2GUPBYjDI","TIj7rkhPpOg","IiJjGkAjU14","9tdOJG4ta1I","PChMxuY6Di0","R-pAakXAWuQ"]

#L'Oreal Voluminous X Fiber Mascara
mascara=["xru4XCteB-o&t","9wT4qvLdyKY",
         'x3FSWaWldAo&t','Dw0OYI1clP0&t','ysqz2S14hW8'
            'F7Yl2cvGbrs&t','OwejOwJreqg','_jzmaPCTJg8']
```

```python
for asin in mascara:
    #extract transcript with api youtube
    c=YouTubeTranscriptApi.get_transcript(asin)
    reviews=[]
    for i in range(len(c)):
        reviews.append(c[i]['text'])
    tokenizer=nltk.tokenize.TreebankWordTokenizer()
    #lowercase the words
    tokens=[]
    for sentence in reviews:
        sentence_lowercase=sentence.lower()
        tokens.append(tokenizer.tokenize(sentence_lowercase))
    #define the words which we want remove from stopwords
    hors_stop_words=['before','after','above','below','under','all','both','few','more','most','some','such','no','nor',
                     'not','so','too','very',"don't",'should',"aren't","couldn't","didn't","doesn't","hadn't","hasn't",
                     "haven't","isn't"]
    stopword_list = nltk.corpus.stopwords.words('english')
    stopword_list=list(stopword_list)
    #add some inutile words to stopwords
    stopword_list.append("always")
    stopword_list.append("already")
    stopword_list.append("hi")
    stopword_list.append("anyway")
    stopword_list.append("currently")
    stopword_list.append("actually")
    stopword_list.append("really")
    stopword_list.append("gon")
    stopword_list.append("na")
    stopword_list.append("thank")
    stopword_list.append("guys")
    stopword_list.append("algebra")
    stopword_list.append("just")
    stopword_list.append("also")
    stopword_list.append("basically")
    stopword_list.append("definitely")
    stopword_list.append("beauties")
    stopword_list.append("channel")
    stopword_list.append("video")
    stopword_list.append("welcome")
    stopword_list.append("everyone")





    #remove important words from stopwords
    for i in hors_stop_words:
        stopword_list.remove(i)
        
        
    #define the tag of each word and let just the adjectives and verbs
    list_words=[]
    for t in tokens:
        listt=[]
        tagged_tokens = nltk.pos_tag(t)
        for t in tagged_tokens:
            if (t[1]=='JJ' or t[1]=='JJS' or t[1]=='JJR' or t[1]=='NN' or t[1]=='NNS' or t[1]=='NNP' or
            t[1]=='RB' or t[1]=='RBR' or t[1]=='RBS' or t[1]=='VB' or t[1]=='VBD' or t[1]=='VBG' or
            t[1]=='VBN'):
                listt.append(t[0])   
        list_words.append(listt)
    
    #clean from stopwords
    without_stop=[]
    for word in list_words:
        listt=[]
        for t in word:
            if t not in stopword_list and t.isalpha():
                listt.append(t)
        without_stop.append(listt)
        
    #frequency of words 
    word_frequencies = {}  
    for t in without_stop:  
        for word in t:
            if word not in word_frequencies.keys():
                word_frequencies[word] = 1
            else:
                word_frequencies[word] += 1
    
    texts=[]
    for word in without_stop:
        texts.append(" ".join(word))
        
    #to form the n grams
    tfidf = TfidfVectorizer(ngram_range=(4,4))
    ff=tfidf.fit_transform(texts)
    #Build a dataset of n gram features
    f=pd.DataFrame(ff.todense(),columns=tfidf.get_feature_names())   
    #extract the weights of those features
    weight_feature=[]
    for column in f:
        listt=[]
        for value in f[column]:
            if value!=0:
                listt.append(value)
        weight_feature.append(listt[0])
    #Display those features
    features=tfidf.get_feature_names()
    #Construct comments from features
    comments=[]
    for i in features:
        comments.append(i)
    #write the data in a csv file
    with open('Comments_youtube_mascara.csv', 'a') as csvFile:
            writer = csv.writer(csvFile)
            writer.writerow(comments)
csvFile.close()
```
